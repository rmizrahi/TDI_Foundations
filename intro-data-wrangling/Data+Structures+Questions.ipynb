4 Create a `set` to identify your favorite hobbies and another `set` to identify someone else's hobbies. Using the two `set`s, what hobbies do you have in common? What hobbies do you not share?

```python
my_hobbies = {'python', 'piano', 'intergalactic space travel'}
your_hobbies = {'python', 'running', 'cooking'}
```

```python
my_hobbies.intersection(your_hobbies)
```

```python
my_hobbies & your_hobbies
```

```python
my_hobbies.difference(your_hobbies)
```

```python
your_hobbies.difference(my_hobbies)
```

```python
my_hobbies.difference(your_hobbies).union(your_hobbies.difference(my_hobbies))
```

```python
(my_hobbies - your_hobbies) | (your_hobbies - my_hobbies)
```

```python
my_hobbies.symmetric_difference(your_hobbies)
```

```python
my_hobbies ^ your_hobbies
```
